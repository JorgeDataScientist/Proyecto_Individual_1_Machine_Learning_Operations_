```python
# Importar librerías
import pandas as pd
import csv
import datetime
from dateutil import parser
```

```python
# Lee el archivo CSV y lo almacena en un DataFrame de Pandas
df = pd.read_csv("datasets_final.csv")
df.head(3)
```

```python
def obtener_nombre_mes(mes):
    nombres_meses = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio',
                     'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']
    return nombres_meses[mes - 1]

def obtener_nombre_dia(dia):
    nombres_dias = ['lunes', 'martes', 'miércoles', 'jueves', 'viernes', 'sábado', 'domingo']
    return nombres_dias[dia]

def procesar_dataframe(df):
    df['release_date'] = pd.to_datetime(df['release_date'])
    df['mes'] = df['release_date'].dt.month.map(obtener_nombre_mes)
    df['dia'] = df['release_date'].dt.dayofweek.map(obtener_nombre_dia)

procesar_dataframe(df)
df.head(10)
```
