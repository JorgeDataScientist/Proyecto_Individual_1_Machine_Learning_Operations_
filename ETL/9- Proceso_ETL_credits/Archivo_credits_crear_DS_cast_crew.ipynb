```python
# Importar librerías
import pandas as pd
```

```python
# Lee el archivo CSV y lo almacena en un DataFrame de Pandas
df = pd.read_csv("crew_data.csv")
df.head(5)
```

```python
# Convertir las columnas numéricas a cadenas
df['crew_credit_id'] = df['crew_credit_id'].astype(str)
df['crew_gender'] = df['crew_gender'].astype(str)
df['crew_id'] = df['crew_id'].astype(str)

# Agrupar los datos por el ID y combinar los valores de las otras columnas
df = df.groupby('id').agg({
    'crew_credit_id': ','.join,
    'crew_department': ','.join,
    'crew_gender': ','.join,
    'crew_id': ','.join,
    'crew_job': ','.join,
    'crew_name': ','.join,
    'crew_profile_path': ','.join
}).reset_index()

# Mostrar el nuevo DataFrame con los datos agrupados
df.head()
```
