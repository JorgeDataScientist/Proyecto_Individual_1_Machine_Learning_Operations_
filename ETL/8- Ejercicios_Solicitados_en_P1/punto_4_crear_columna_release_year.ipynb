```python
# Importar librerías
import pandas as pd
```

```python
# Lee el archivo CSV y lo almacena en un DataFrame de Pandas
df = pd.read_csv("../DataSet_Limpio/dataSet_limpio.csv")
df.head(3)
```

### Revision de columna:

De haber fechas, deberán tener el formato AAAA-mm-dd

```python
# Convertir la columna "release_date" en tipo datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Contar la cantidad de valores no nulos en la columna "release_date"
cantidad_formato_valido = df['release_date'].notnull().sum()

# Imprimir la cantidad de valores con formato válido
print("Cantidad de valores con formato válido en la columna 'release_date':", cantidad_formato_valido)
```

### Crear la columna release_year donde extraerán el año de la fecha de estreno

```python
# Extraer el año de la columna "release_date" y crear la columna "release_year"
df['release_year'] = df['release_date'].dt.year

# Imprimir el DataFrame con la nueva columna "release_year"
df.head(3)
```
